# Seaborn

```python
import matplotlib.pyplot as  plt
import seaborn as sns
import pandas as pd 
```

```python
df = pd.read_csv('pizza.csv')
df.head()
```

```python
df['total_price'].sum()
```

```python
df['total_price'].isnull().sum()
```

```python
sns.lineplot(x='pizza_id',y='unit_price',data=df.head(10))
plt.show()
```

```python
plt.figure(figsize=(8,6))
sns.barplot(x='pizza_id',y='unit_price',data=df.head(10), hue='pizza_name',palette="Set3",legend=False)
plt.show()
```

```python
avg = df['pizza_name'].value_counts()
```

```python
pizza_counts = df['pizza_name'].value_counts()
total_pizzas = len(df)
pizza_share = pizza_counts / total_pizzas
pizza_stats = pd.DataFrame({'Pizza Name': pizza_counts.index, 'Count': pizza_counts.values, 'Share': pizza_share.values})
pizza_stats
```

```python
labels = pizza_stats['Pizza Name'].head()
sizes = pizza_stats['Share'].head()
largest_stats = sizes.idxmax()   # find the index of max share
explode = [0.1 if i==largest_stats else 0 for i in range(len(sizes))] # explode = [0.1,0,0,0,0]

plt.pie(
    sizes,                  # 1. sizes: An array-like object specifying the sizes of each slice.
    labels=labels,          # 2. labels: An optional list of labels for each slice (default is None).
    autopct='%1.1f%%',      # 3. autopct: A format string for displaying percentages on each slice.
    startangle=90,          # 4. startangle: The angle at which the first slice starts (default is 0 degrees).
    shadow=False,           # 5. shadow: Whether to add a shadow effect to the pie chart (default is False).
    explode=explode,           # 6. explode: An array-like object specifying the fraction of the radius by which to offset each slice (default is None).
    pctdistance=0.6,        # 7. pctdistance: The distance of the percentage label from the center (default is 0.6).
    colors=None,            # 8. colors: An optional list of colors for each slice (default is None).
    wedgeprops=None,        # 9. wedgeprops: Dictionary of properties for the wedge objects (e.g., edgecolor, linewidth).
    textprops=None,         # 10. textprops: Dictionary of properties for the text label objects (e.g., fontsize, color).
    center=(0, 0),          # 11. center: Coordinates of the center of the pie chart (default is (0,0)).
    frame=False,            # 12. frame: Whether to draw a frame around the pie chart (default is False).
    rotatelabels=False,     # 13. rotatelabels: Whether to rotate the labels to align with the slices (default is False).
    radius=1.0             # 14. radius: The radius of the pie chart (default is 1.0).
)

plt.axis('equal')           # Ensure the pie chart is drawn as a circle.

# Add any additional customization, such as a title or legend, before plt.show()
```

```python
labels = pizza_stats['Pizza Name'].head()
sizes = pizza_stats['Share'].head()
plt.pie(sizes, 
        labels=labels, 
        autopct='%.2f%%', 
        startangle=90,
        wedgeprops={'edgecolor': 'yellow'}, 
        textprops={'fontsize': 10,'color':'#eb6734'}
       )
plt.axis('equal')
plt.show()
```
