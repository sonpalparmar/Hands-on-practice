```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('Airtel.csv')
df.head()
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df.describe()
```

```python

plt.figure(figsize=(10, 6))
sns.lineplot(x='VALIDITY', y='OFFER_PRICE', data=df)
plt.title('Line Plot of VALIDITY vs.OFFER_PRICE for Airtel')
plt.xlabel('VALIDITY')
plt.ylabel('OFFER_PRICE')
plt.show()
```

```python
heat_df = df.drop('SIM_COMPANY',axis=1)
correlation_matrix = heat_df.corr()

plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', linewidths=0.5)
plt.title('Correlation Heatmap')
plt.show()
```

```python
sns.pairplot(df, hue='VALIDITY', palette='Set1')
plt.show()
```
